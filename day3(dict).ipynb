# **Dictionary in Python**

**dictionary is a key:value pair pairs inside curly brackets {}**

**Notes**


*   Dictionary keys must be immutable, such as tuples, strings, integers, etc. We cannot use mutable (changeable) objects such as lists as keys.

*   Keys of a dictionary must be unique

dictionary unpacking:


````python
# Creating a dictionary
my_dict = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Accessing values
print(my_dict["name"])  # Output: John
````


**Modifying and Adding Elements:**


````python
# Modifying and adding elements
my_dict["age"] = 35
my_dict["gender"] = "Male"
````

**mostly used dict mehtods**



````python
# Using dictionary methods
print(my_dict.keys())    # Output: dict_keys(['name', 'age', 'city', 'gender'])
print(my_dict.values())  # Output: dict_values(['John', 35, 'New York', 'Male'])
print(my_dict.items())   # Output: dict_items([('name', 'John'), ('age', 35), ('city', 'New York'), ('gender', 'Male')])
````

````python
# nested dict
employees = {
    "E101": {"name": "John Doe", "age": 30, "department": "HR"},
    "E102": {"name": "Jane Smith", "age": 35, "department": "Finance"},
    "E103": {"name": "Tom Brown", "age": 28, "department": "IT"}
}
````

**Dictionary Comprehension**
```
Dictionary comprehension is a concise and elegant way to create dictionaries in Python using a single line of code


````python
{key_expression: value_expression for item in iterable}
````

````python
# Dictionary comprehension to square numbers
squared_dict = {num: num ** 2 for num in range(1, 6)}
print(squared_dict)
````
